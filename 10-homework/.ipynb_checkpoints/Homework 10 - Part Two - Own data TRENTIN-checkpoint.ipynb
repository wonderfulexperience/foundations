# 6. Descriptions of four columns
Bullish: Ratio of investors which are bullish

Total: Check-sum if percentages add to 1 (100%)

Bull-Bear Spread: Bullish minus Bearish ratio

S&P 500 Weekly Close: Last quote for the S&P 500 stock index in that week

Definitions are made up. This is the official definition:

<i>The AAII Investor Sentiment Survey measures the percentage of individual investors who are bullish, bearish, and neutral on the stock market for the next six months; individuals are polled from the ranks of the AAII membership on a weekly basis. Only one vote per member is accepted in each weekly voting period.</i>

# 7. Five questions to ask the data
a. How did the average investor sentiment change by decade?<br>
b. What were the fastest changes in investor sentiment?<br>
c. Is there a relationship between investor sentiment changes and subsequent stock price changes?<br>
d. Has investor sentiment become more volatile over time?<br>
e. Is there a seasonal pattern for investor sentiment?

# 8. Answers

```python
# a. How did the average sentiment change by decade?

# create decade variable

df.loc[df.Date.str[:3] == '198', 'decade'] = "Eighties"
df.loc[df.Date.str[:3] == '199', 'decade'] = "Nineties"
df.loc[df.Date.str[:3] == '200', 'decade'] = "2000s"
df.loc[df.Date.str[:3] == '201', 'decade'] = "2010s"
(df.groupby('decade')['Bull-Bear Spread'].mean()*100)\
    .sort_values().plot(kind='barh', figsize = (10,10))
    
# The Nineties and 2000s showed almost the same level of very optimistic.
# The 2010s show a much less bullish sentiment. The Eighties were the worst.
```

```python
# b. What were the fastest changes in investor sentiment?
# (using four-weeks changes in percentage points)

sentiment_change = \
    (df['Bull-Bear Spread'] - df['Bull-Bear Spread'].shift(-4))*100
df['sentiment_change_4w'] = sentiment_change

print("Largest negative changes")
print(df[['Date','sentiment_change_4w']]
      .sort_values('sentiment_change_4w').head(5))
print("\nLargest positive changes")
print(df[['Date','sentiment_change_4w']]
      .sort_values('sentiment_change_4w', ascending=False).head(5))
```

```python
#c. Is there a relationship between investor sentiment changes
# and subsequent stock price changes?
# - using graphical analysis

#take inverse value of pct_change as time order is reversed
subsequent_stock_change = ((1/(df['S&P 500 Weekly Close']
                               .pct_change(periods = 1)+1))-1)*100
df['stock_change'] = subsequent_stock_change
df[['Date','stock_change','S&P 500 Weekly Close']]

sentiment_change = \
    (df['Bull-Bear Spread'] - df['Bull-Bear Spread'].shift(-1))*100
df['sentiment_change'] = sentiment_change

df.plot(kind='scatter', x='sentiment_change', y='stock_change')
```

```python
# doesn't look like any strong relationship between changes
# what about absolute values of sentiment and subsequent four week changes?

df['stock_change_4w']=(df['S&P 500 Weekly Close']
                       .shift(4) - df['S&P 500 Weekly Close'])\
                        /df['S&P 500 Weekly Close']*100

df.plot(kind='scatter', x='Bull-Bear Spread', y='stock_change_4w')
```

```python
# trying out some linear regression (ordinary least squares, OLS)

import matplotlib.pyplot as plt
import statsmodels.regression.linear_model as sm

x = df['Bull-Bear Spread']
x = sm.add_constant(x)
regression = sm.OLS(df['stock_change_4w'], x, missing='drop').fit()
regression.summary()

# no statistical evidence at all (R-squared is 0)
```

```python
# d. Has investor sentiment become more volatile over time?

# plot rolling 1 year standard deviation of investor sentiment
# the order of dates has to be reversed


std_sentiment = df.sort_values('Date', ascending=True)['Bull-Bear Spread'].rolling(window=52).std()
df['std_sentiment'] = std_sentiment
```

```python
df.dropna().plot('Date','std_sentiment',figsize = (10,10)).invert_xaxis()

# in the last years a steady decline of the standard deviation of investor sentiment
# --> sentiment became less volatile since the year 2003
```

```python
# e. Is there a seasonal pattern for investor sentiment?

# showing in one graph the development of investor sentiment in each year

import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(18, 8))

this_color = 0.8

for this_year in range(1988,2017): #only full years
    this_df = df[df['year']==str(this_year)].dropna()
    if this_df['Date'].count() > 51: #only years with 52 weeks or more
        this_index = range(1, this_df['Bull-Bear Spread'].count()+1)
        
        #chart each year on top of each other
        ax.plot(this_index, this_df['Bull-Bear Spread'], color=str(this_color), linewidth=0.5)
        this_color -= 0.02
    
# building a mean for each week

year_list = []

for this_year in range(1988,2017):
    this_df = df[df['year']==str(this_year)].dropna()
    
    weeks_list = []
    
    count = 1
    if this_df['Date'].count() > 51: #only years with 52 weeks or more
        for index, row in this_df.iterrows():
            sentiment = float(row['Bull-Bear Spread'])
            weeks_list.append(sentiment)
            count += 1
            if count == 53: # do the loop until 52 weeks are full
                break
       
    #check that no empty week is added
    
    if weeks_list:
        year_list.append(weeks_list)

#create list for averages of each week
week_values = []
    
for this_week in range(0,52):
    this_week_list = []
    
    for this_year in range(0,len(year_list)):
        this_week_list.append(year_list[this_year][this_week])
        
    this_week_mean = np.mean(this_week_list)
    week_values.append(this_week_mean)

this_index = range(1,53)

#chart the mean values for each week
plt.plot(this_index, week_values, color = 'red', linewidth=2.5)
plt.show()

#On average, investor sentiment seems to be a slight decrease in the beginning of the year
```
